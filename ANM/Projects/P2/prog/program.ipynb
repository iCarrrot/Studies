Używam funkcji interpol($\epsilon$, czyZpliku,nazwa,sposob,typ),   gdzie: 

$\epsilon$ to odległość pomiędzy kolejnymi punktami gotowej krzywej sklejanej, \\
czyZpliku przyjmuje wartość 1 (odczyt z pliku) lub 0 (wybór zaimplementowanej funkcji matematycznej); \\
nazwa to nazwa pliku (gdy czyZpliku=1) lub żądanej funkcji matematycznej (w przeciwnym przypadku);\\
sposob to sposób tworzenia t (''pitagoras'',''równoodległe'',''czebyszew'');\\
typ to typ funkcji (''okresowa'',''naturalna'',''wielomian'').\\


```julia
(X,Y)=interpol(0.001,1,"polska.pkt","pitagoras","okresowa")
plot(X, Y, label="Krzywa", "red", linewidth=1.0 )

axis("equal")
```

```julia
(X,Y)=interpol(0.0001,1,"marchewka.pkt","czebyszew","wielomian")

plot(X, Y, label="Krzywa", "blue", linewidth=1.0 )
(X1,Y1)=(X,Y)=interpol(0.001,1,"marchewka.pkt","pitagoras","okresowa")
plot(X1, Y1, label="Krzywa", "red", linewidth=1.0 )

axis("equal")
axis([-10,23,-5,20])
```

```julia
(X,Y)=interpol(0.0001,1,"marchewka.pkt","czebyszew","wielomian")

plot(X, Y, label="Krzywa", "blue", linewidth=1.0 )
(X1,Y1)=(X,Y)=interpol(0.001,1,"marchewka.pkt","pitagoras","naturalna")
plot(X1, Y1, label="Krzywa", "red", linewidth=1.0 )

axis("equal")
axis([-10,23,-5,20])
```

```julia
ε=0.00001
(X,Y)=interpol(ε,0,"okrąg","pitagoras","okresowa")
X1=Float64[]
Y1=Float64[]
iter=0.0
println(length(X))
X2=linspace(0,2*pi,length(X));Y2=sin.(X2);
for i in 1:length(X)
    push!(Y1,-log10(abs(Float64(Y2[i]-Y[i])))/8)
    push!(X1,X[i])
    #println(i," ",Y1[i])
end
plot(X1,Y1, linewidth=0.3)

plot(X,Y,"red",linewidth=2.0)
t=linspace(0,2*pi-pi/100,1000);x=cos.(t);y=sin.(t);
plot(x,y,"black",linewidth=0.4)
axis("equal")
    

```
